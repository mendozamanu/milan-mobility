```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import cross_val_score, cross_val_predict
#from sklearn.cluster import KMeans, OPTICS
```

```python
#Week 4-10 nov: smsin traffic
#data = pd.read_csv("./classif/randomgrid/classif-randomgridweek_smsin.csv")
#data = pd.read_csv("./classif/weekly10x10/classif10x10week_smsin.csv")
data = pd.read_csv("./classif/20x20/classif20x20week_smsin.csv")
data.head()
```

```python
y = data.iloc[:,170] #h/w
x = data.iloc[:,2:170] #smsin for the whole week (24x7)
```

```python
### Clustering ###
#model = MeanShift()
#model = KMeans(n_clusters=2)
#model = OPTICS(min_samples=3)
# fit model and predict clusters
#yhat=model.fit_predict(x)
# retrieve unique clusters
#clusters = np.unique(yhat)
#print(len(yhat))
###            ###
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
scores1 = cross_val_score(knn, x, y, cv=5)
scores2 = cross_val_score(knn, x, y, cv=10)
```

```python
print("5 fold: %0.2f accuracy with a standard deviation of %0.2f" % (scores1.mean(), scores1.std()))
print("10 fold: %0.2f accuracy with a standard deviation of %0.2f" % (scores2.mean(), scores2.std()))
```
