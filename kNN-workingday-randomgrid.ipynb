```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
data = pd.read_csv("./classif/randomgrid/classif-randomgrid_smsin.csv")
#data = pd.read_csv("./classif/randomgrid/classif-randomgridmonminute_smsin.csv")
data.head()
#random grid, monday(4/11) - smsin
```

```python
y = data.iloc[:,26] #h/w
x = data.iloc[:,2:26] #sms, calls, internet
#y = data.iloc[:,146] #h/w
#x = data.iloc[:,2:146] #sms, calls, internet
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
scores1 = cross_val_score(knn, x, y, cv=5)
scores2 = cross_val_score(knn, x, y, cv=10)
```

```python
print("5 fold: %0.2f accuracy with a standard deviation of %0.2f" % (scores1.mean(), scores1.std()))
print("10 fold: %f accuracy with a standard deviation of %f" % (scores2.mean(), scores2.std()))
```
