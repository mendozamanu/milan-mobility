```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
data = pd.read_csv("./classif/fullgrid/classifgridmon_callout.csv")
data2 = pd.read_csv("./classif/20x20/classif20x20_callout.csv")
#data = pd.read_csv("./classif/20x20/classif20x20frid_callout.csv")
#data = pd.read_csv("./classif/20x20/classif20x20monminute_callout.csv")

data2.head()
#20x20 grid, monday(4/11) - callout / friday(8/11)
```

```python
y = data2.iloc[:,26] #h/w
x = data2.iloc[:,2:26] #sms, calls, internet
#y = data.iloc[:,146] #h/w
#x = data.iloc[:,2:146] #sms, calls, internet
x_test = data.iloc[:,2:26]
```

```python
x_test.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x,y)
scores1 = cross_val_score(knn, x, y, cv=5)
scores2 = cross_val_score(knn, x, y, cv=10)
```

```python
print("5 fold: %0.2f accuracy with a standard deviation of %0.2f" % (scores1.mean(), scores1.std()))
print("10 fold: %0.2f accuracy with a standard deviation of %0.2f" % (scores2.mean(), scores2.std()))
```

```python
y_predt = knn.predict(x_test) #10
```

```python
predicted = pd.DataFrame(np.transpose(y_predt),columns=['predicted'])
predicted['cellid']=data.cellid
#predicted.append(np.transpose(y_predt))
predicted.head()
```

```python
predicted.to_csv('./csv/predicted_full.csv')
```
